```python
import os
from glob import glob
import operator
from functools import reduce
```

```python
# input - original (complete dataset - where to copy images from) dataset path
# io_p = '../datasets/helen-mirrored-new/output/trainset/txtmaps-hstack'
io_p = '/media/dataserver/workspace/blanca/training_datasets/pix2pix/images_target_clean_classified/good_fit'

# input - clean dataset path
# ic_p = '../datasets/clean-txtmap-dataset/txtmaps-hstack_helen_train'
ic_p = '/media/dataserver/workspace/blanca/training_datasets/classifier/training/good_fit'

# output - where to copy images path
# od_p = '../datasets/clean-txtmap-dataset/txtmaps-hstack_helen_train-rb'
od_p = '/media/dataserver/workspace/blanca/training_datasets/classifier/training/bad_fit'

i_ext = ['*.jpg', '*.JPG', '*.png']

io_pl = reduce(operator.add, [glob(os.path.join(io_p, i)) for i in i_ext], [])
ic_pl = reduce(operator.add, [glob(os.path.join(ic_p, i)) for i in i_ext], [])

io_nl = [i.split('/')[-1] for i in io_pl]
ic_nl = [i.split('/')[-1] for i in ic_pl]

od_nl = list(set(io_nl) - set(ic_nl))

print('Total (original): %s | Clean: %s | Deleted: %s' %('{:,}'.format(len(io_nl)), '{:,}'.format(len(ic_nl)), '{:,}'.format(len(od_nl))))

od_pl = [i for i in io_pl if i.split('/')[-1] not in ic_nl]
print(len(od_pl))
```

```python
[os.system('cp %s %s' %(i, od_p)) for i in od_pl]
```
